Deriving the Characteristic Equation

$Ax = \lambda x$

$A = $ Linear transformation vector

$x = $ Eigenvector

$\lambda = $ Eigenvalue (scalar value)

$Ax - \lambda x = 0$

$Ax - \lambda I x = 0$

$(A - \lambda I)x = 0$

This equation has non-zero solution only if $(A - \lambda I)$ is zero and $|A - \lambda I| = 0$ (determinant)

This polynomial function of the variable $\lambda$ has degree $n$, which is the order of $A$ (rows multiplied by columns)

These principles are the foundation of dimensionality reduction for ML algorithms using PCA

Speeding up raising a matrix to a power

```python
A = np.matrix([[1, 4],
               [9, 1]])
A
```

```python
eigvals, eigvecs = eig(A)
```

```python
eigvec1 = eigvecs[:,0]
eigvec2 = eigvecs[:,1]

eigvec1, eigvec2
```

```python
lambda1 = eigvals[0]
lambda2 = eigvals[1]

lambda1, lambda2
```

```python
P = np.array([eigvec1, eigvec2]).T
P
```

```python
D = np.diag((lambda1, lambda2))
D
```

```python
from scipy.linalg import inv
```

```python
M = P @ D @ inv(P)
M
```

```python
A
```

```python
np.allclose(M, A)
```

```python
k = 3
```

```python
P @ D**k @ inv(P)  # only the diagonal matrix needs to be raised to the power of k
```

```python
result = A.copy()

for _ in range(1, k):
  result = result @ A

print(result)
```

```python
%%timeit

P @ D**k @ inv(P)
```

```python
%%timeit

result = A.copy()

for _ in range(1, k):
  result = result @ A
```

SVD on an image

```python
import os
import numpy as np
from numpy.linalg import svd
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
```

```python
img = plt.imread('/content/apex_epa_example.png')

plt.figure(figsize=(14, 8))

plt.imshow(img)
plt.show()
```

```python
img[0:2]
```

```python
img.shape
```

```python
gray_img = rgb2gray(img)
plt.figure(figsize=(12,8))
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
```

```python
gray_img.shape
```

```python
gray_img[:2]
```

```python
u, s, v = svd(gray_img)
```

```python
u[:5]
```

```python
v[:5]
```

```python
u.shape, s.shape, v.shape
```

```python
comps = [comp for comp in range(u.shape[0], 0, -60)]

comps
```

```python
comps = np.sort(comps)

comps
```

```python
u_sub = u[:, :comps[0]]

u_sub
```

```python
u_sub.shape
```

```python
s_sub = s[:comps[0]]

s_sub
```

```python
s_sub.shape
```

```python
s_sub_diag = np.diag(s_sub)

s_sub_diag
```

```python
s_sub_diag.shape
```

```python
v_sub = v[:comps[0], :]

v_sub
```

```python
v_sub.shape
```

```python
reconstructed_img = u_sub @ s_sub_diag @ v_sub

reconstructed_img.shape
```

```python
plt.figure(figsize=(14, 8))
plt.imshow(reconstructed_img, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(20, 12))

for i in range(len(comps)):

  reconstructed_img = u[:, :comps[i]] @ np.diag(s[:comps[i]]) @ v[:comps[i], :]

  if (comps[i] == u.shape[0]):

    plt.subplot(4, 3, i+1)
    plt.imshow(reconstructed_img, cmap='gray')
    plt.axis('off')
    plt.title('Original with n_components =' + str(comps[i]))

  else:
    plt.subplot(4, 3, i+1)
    plt.imshow(reconstructed_img, cmap='gray')
    plt.axis('off')
    plt.title('n_components =' + str(comps[i]))
```

```python
comp = 314

plt.figure(figsize=(12, 8))

reconstructed_img = u[:, :comp] @ np.diag(s[:comp]) @ v[:comp, :]

plt.imshow(reconstructed_img, cmap='gray')

plt.axis('off')
```
