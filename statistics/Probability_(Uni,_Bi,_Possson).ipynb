```python
val, cnt = np.unique(outcomes, return_counts=True)

prop = cnt/len(outcomes)
```

## Poisson

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import poisson
```

```python
poisson.pmf(k=8, mu=10)  # probability of exactly 8 given average of 10
```

```python
poisson.pmf(k=10, mu=10)
```

```python
range = np.arange(0,20)
```

```python
poisson_dist = poisson.pmf(k=range, mu=10)
poisson_dist
```

```python
plt.figure(figsize=(12, 8))
plt.plot(range, poisson_dist, '-o', label='mu = 10')

plt.xlabel('# Response')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.legend()
plt.show()
```

```python
sum(poisson_dist)
```

```python
range = np.arange(0, 150)
```

```python
mu_10 = poisson.pmf(k=range, mu=10)
mu_20 = poisson.pmf(k=range, mu=20)
mu_50 = poisson.pmf(k=range, mu=50)
mu_100 = poisson.pmf(k=range, mu=100)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(range, mu_10, label='mu = 10')
plt.plot(range, mu_20, label='mu = 20')
plt.plot(range, mu_50, label='mu = 50')
plt.plot(range, mu_100, label='mu = 100')

plt.xlabel('# Response')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.legend()
plt.show()
```
