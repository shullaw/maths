# Linear functions cont'd

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def position_func(t):  # distance given time
  return 20*t + 5      # linear
```

```python
t = np.linspace(0, 10, 100)
t
```

```python
plt.figure(figsize=(12,6))

plt.plot(t, position_func(t),
          color='purple',
          lw=3,
          label='Position graph')

plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')

plt.legend(loc='upper center')

plt.xlabel("Time (in hours)")
plt.ylabel("Distance covered (in km)")

plt.grid(True)
```

```python
def get_derivative(f, t:np.array, method:str='central', h:float=.0001):
  
  if method == 'central':
    return (f(t + h) - f(t - h)) / (2*h)

  elif method == 'forward':
    return (f(t + h) - f(t)) / h

  elif method == 'backward':
    return (f(t) - f(t - h)) / h

  else:
    raise ValueError("Method must be 'central', 'forward' or 'backward'.")
```

```python
velocity_5 = get_derivative(position_func, 5)
velocity_5
```

```python
velocity_10 = get_derivative(position_func, 10)
velocity_10
```

```python
from scipy.misc import derivative
```

```python
derivative(position_func, 5)
```

```python
derivative(position_func, 10)
```

```python
def get_velocity(t):
  return derivative(position_func, t)
```

```python
t_1 = 4
```

```python
plt.figure(figsize=(12,6))

plt.plot(t, position_func(t),
          color='purple',
          lw=3,
          label='Distance graph')

plt.plot(t, get_velocity(t),
         color='green',
         lw=3,
         label='Velocity graph')

plt.scatter(t_1, derivative(position_func, t_1),
            color='red',
            lw=5,
            label=r'$Velocity/ at/ Time/ (t_1) = 4/ hours$')

plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')

plt.legend(loc='upper center')

plt.xlabel("Time (in hours)")
plt.ylabel("Distance covered (in km)")

plt.grid(True)
```

# Quadratic and Polynomial Functions

```python
x = np.linspace(-10, 10, 100)
```

```python
def f(x):
  return x**2 + 0*x + 0  # ax^2 + bx + c
```

```python
plt.figure(figsize=(12,6))

plt.plot(x, f(x),
         color='purple',
         lw=2,
         label='Error')

plt.xlim(-12, 12)
plt.ylim(0, 100)

plt.xlabel('x')
plt.ylabel('Error')
plt.legend(loc='upper center')
plt.grid(True)
plt.show()
```

```python
def tangent(x, x1, y1):
  
  m = derivative(f, x1)  # geometric interpretation of dy/dx

  return m * (x - x1) + y1  # 2-point form, y1 - y / x1 - x
```

```python
x1 = 5
y1 = f(x1)
```

```python
print(f'Rate of change of error at {x1, y1} = {derivative(f, x1)}')
```

```python
# define t range for tangent
x_range = np.linspace(x1 -5, x1 + 5, 100)
```

```python
plt.figure(figsize=(12,6))

plt.plot(x, f(x),
         color='purple',
         lw=2,
         label='Error')

plt.scatter(x1, y1, color='green', s=50)

plt.plot(x_range, tangent(x_range, x1, y1),
         linestyle='--',
         color='blue',
         lw=2,
         label='tangent')

plt.xlim(-10, 10)
plt.ylim(0, 100)

plt.xlabel('x')
plt.ylabel('Error')
plt.legend(loc='upper center')
plt.grid(True)
plt.show()
```

# Partial Derivatives (hold all but one variable constant)

Partial derivative of $f$ with respect to $x$

$f(x, y) = 2x^3 + 3y^3$

$f(x, 2) = 2x^2 + 24$

$\frac{∂(x,y)}{\partial x} = \frac{\partial(2x^2 + 24)}{\partial x}$

$\frac{\partial f(x,y)}{\partial x} = 4x$

Partial derivative of $f$ with respect to $y$
$f(x, y) = 2x^3 + 3y^3$

$f(3, y) = 18 + 3y^3$

$\frac{∂(x,y)}{\partial y} = 9y^2$ (Derivative of constant is 0)




Partial derivative of cylinder with respect to radius
$V = \pi r^2 h$

$h = H$

$\frac{\partial V}{\partial x} = \frac{\partial \pi r^2 H}{\partial x}$

$\frac{\partial V}{\partial r} = 2\pi rH$

```python
import math
import numpy as np

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

from scipy.misc import derivative
```

```python
@np.vectorize
def func(x, y):  # function to differentiate 
  return x**2 + y**2 + x*y
```

```python
plt.figure(figsize=(14,8))

ax = plt.axes(projection='3d')

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z = func(X, Y)  # value of function

ax.plot_surface(X, Y, Z,
                color='purple',
                antialiased=True,
                edgecolor='orange')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
```

```python
def get_partial_derivative(func, const_value, x=None, y=None):

  h=.0001

  # func with respect to x
  if x is not None:
    y = const_value
    return (func(x + h, y) - func(x, y)) / h

  # func with respect to y
  if y is not None:
    x = const_value  
    return (func(x, y+h) - func(x, y)) / h
    # partial derivatives requires other 

    # variables to remain constant
    # numerators are difference between outputs of function
```

```python
const_value = 1
```

```python
get_partial_derivative(func, const_value, x=1)
```

```python
get_partial_derivative(func, const_value, x=5)
```

```python
get_partial_derivative(func, const_value, y=1)  # func is symmetric
```

```python
get_partial_derivative(func, const_value, x=5)
```

```python
plt.figure(figsize=(14,8))

ax = plt.axes(projection='3d')

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z0 = func(X, Y)  # value of function

ax.plot_surface(X, Y, Z0,
                color='purple',
                antialiased=True,
                edgecolor='orange')

Z1 = get_partial_derivative(func, const_value, x=X)  
Z2 = get_partial_derivative(func, const_value, y=Y)

ax.plot_surface(X, Y, Z1,
                color='cyan',
                antialiased=True,
                edgecolor='blue')

ax.plot_surface(X, Y, Z2,
                color='black',
                antialiased=True,
                edgecolor='red')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
```
